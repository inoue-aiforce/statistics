```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10, 10, 0.1)
sample_size = 200000
```

```python
f_cauchy = 1 / (np.pi * (1 + x**2))

plt.plot(x, f_cauchy)
plt.ylabel("f (x)")
plt.xlabel("x")
```

```python
X_cauchy = np.random.standard_cauchy(sample_size)
plt.hist(X_cauchy, bins=100)
plt.xlabel("X")
plt.ylabel("frequency")
```

```python
plt.plot(np.add.accumulate(X_cauchy)/range(sample_size))
plt.ylabel("f (x)")
plt.xlabel("x")
```

```python
f_cauchy_expected = x*(1 / (np.pi * (1 + x**2)))
```

```python
fig,ax = plt.subplots(figsize=(4,3),facecolor='white',dpi=150)
ax.plot(x,f_cauchy_expected,marker='.')

ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(direction='inout')

plt.show()
```

```python
f_normal =  np.exp(-x**2 / 2) / np.sqrt(2 * np.pi)

plt.plot(x, f_normal)
plt.ylabel("f (x)")
plt.xlabel("x")

```

```python
X_normal = np.random.randn(sample_size)
plt.hist(X_normal, bins=100)
plt.xlabel("X")
plt.ylabel("frequency")
```

```python
plt.plot(np.add.accumulate(X_normal)/range(sample_size))
plt.ylabel("f (x)")
plt.xlabel("x")
```

```python
f_normla_expected = x*(np.exp(-x**2 / 2) / np.sqrt(2 * np.pi))
```

```python
fig,ax = plt.subplots(figsize=(4,3),facecolor='white',dpi=150)
ax.plot(x,f_normla_expected,marker='.')

ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(direction='inout')

plt.show()
```
